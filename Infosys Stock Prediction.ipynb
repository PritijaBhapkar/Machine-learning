```python
moving_avg = df['Close'].rolling(window=20).mean()

candlestick = go.Candlestick(x=df.index,        
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'],
                name='Candlestick')

moving_avg_trace = go.Scatter(x=df['Date'],
                             y=moving_avg,
                             mode='markers',
                             line=dict(color='blue', width=4))

fig = go.Figure(data=[candlestick, moving_avg_trace])

fig.update_layout(title='Moving avg line',
                   xaxis_title='Date',
                   yaxis_title='Price')

fig.show()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
```

```python
X = np.array(df.index).reshape(-1,1)
Y = df['Close']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

```python
scaler = StandardScaler().fit(X_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# Plot a grpah

trace0 = go.Scatter(
     x = X_train.T[0],
     y = Y_train,
     mode = 'markers',
     name = 'Actual'
)
trace1 = go.Scatter(
     x = X_train.T[0],
     y = regressor.predict(X_train).T,
     mode = 'lines',
     name = 'Predicted'
)
df_data = [trace0, trace1]
layout.xaxis.title.text = 'Day'
plot2 = go.Figure(data = df_data, layout = layout)
```

```python
pio.show(plot2)
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
r2 = r2_score(Y_test, Y_pred)
print("R-squared score:", r2)
```
