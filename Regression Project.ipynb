```python
cols = ['Vehicle Class', 'Engine Size(L)', 'Cylinders', 'Fuel Type', 'Transmission', 'Fuel Consumption Comb (L/100 km)']
X = df[cols]
y = df[['CO2 Emissions(g/km)']]     # why 2 brackets -> so it will extracted in 2d format
```

```python
# Independent features
X.head()
```

```python
# dependent
y.head()
```

```python
print('Indepenedent feature set shape :', X.shape)
print('Dependent feature set shape :', y.shape)
```

### Splitting data into training and test sets


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print('Training data shape: ',X_train.shape)
print('Training labels shape: ',y_train.shape)
print('Testing shape: ', X_test.shape)
print('Testing labels shape: ',y_test.shape)
```

## Scaling:    
    Used to bring data to same scale

#### Standard Scaler
    X_scaled = (X - X_mean) / X_std
Performs z score normalization
Zero mean
Unit variance
#### MinMax Scaler
    X_scaled = (X - X_min) / (X_max - X_min)
Limits the data between 0 to 1

```python
cols = [ 'Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fitting on training set
scaler.fit(X_train[cols])

# Transform training and test sets
X_train[cols] = scaler.transform(X_train[cols])
X_test[cols] = scaler.transform(X_test[cols])
```

## Encoding (performed for categorical data)
   --> Numerical representation to categorical data

```python
cols = ['Fuel Type','Transmission','Vehicle Class']
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

# Fitting on training set
encoder.fit(X_train[cols])

# Transform training and test sets
X_train[cols] = encoder.transform(X_train[cols])
X_test[cols] = encoder.transform(X_test[cols])
```

```python
X_train.head(10)
```

 ### Training the ML model

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors = 3)
regressor.fit(X_train, y_train)
```

### Evaluating the model

```python
pred_train = regressor.predict(X_train)
pred_test  = regressor.predict(X_test) 
 
from sklearn.metrics import r2_score
print('Training Set Score : ', r2_score(y_train, pred_train))
print('Testing Set Score : ', r2_score(y_test, pred_test))
```
